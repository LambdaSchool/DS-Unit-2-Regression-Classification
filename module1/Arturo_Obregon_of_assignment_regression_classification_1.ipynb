```python
# Read New York City apartment rental listing data
import pandas as pd
df = pd.read_csv('../data/apartments/renthop-nyc.csv')
assert df.shape == (49352, 34)
```

```python
# Remove outliers: 
# the most extreme 1% prices,
# the most extreme .1% latitudes, &
# the most extreme .1% longitudes
df = df[(df['price'] >= 1375) & (df['price'] <= 15500) & 
        (df['latitude'] >=40.57) & (df['latitude'] < 40.99) &
        (df['longitude'] >= -74.1) & (df['longitude'] <= -73.38)]
```

```python
df.head(10)
```

```python
df_copy = df[df['bedrooms'] == 4]
```

```python
# df_copy.head()
df_copy_copy = df_copy[df_copy['bathrooms']==3]
df_copy_copy.sort_values('price', axis=0, ascending=False,inplace=True)
df_copy_copy.head(20)
```

```python
df.describe()
```

```python
import plotly.express as px
px.scatter(df, x='bedrooms', y='longitude')
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

features = ['bedrooms']
target = 'price'
X_train = df[features]
Y_train = df[target]

model.fit(X_train, Y_train)
```

```python
bedrooms = 3
X_test = [[bedrooms]]

model.predict(X_test)
```

```python
y_pred = model.predict(X_test)
print(f'Predicted price for {X_test} bedrooms in NY' + str(y_pred))
```

```python
y_test = [5000]
```

```python
model
```

```python
m = model.coef_[0]
b = model.intercept_
print('y-=mx+b')
print(f'y = {m: .0f}* + {b: .0f}')
print(f'price  = {m: .0f}*square_feet + {b: .0f}')
```

```python
def predict(bedrooms):
    y_pred = model.predict([[bedrooms]])
    estimate = y_pred[0]
    coefficient = model.coef_[0]
    result = f'${estimate:,.0f} esitmated price for {bedrooms:,.0f} bedroom condo in New York. '
    explanation = f'In this linear regression, each additional bedroom adds ${coefficient:,.0f}.'
    return result + explanation
  
predict(2)
```

```python
from sklearn.linear_model import LinearRegression

model_multi = LinearRegression()

features2 = ['bedrooms', 'bathrooms']
target2 = 'price'
X_train2 = df[features2]
Y_train2 = df[target2]

model_multi.fit(X_train2, Y_train2)
```

```python
bedrooms2 = 4
bathrooms2 = 3
X_test2 = [[bedrooms2, bathrooms2]]

model_multi.predict(X_test2)
```

```python
y_pred_3 = model_multi.predict(X_test2)
print(f'Predicted price for {X_test2} bedroom/bathroom in NY' + str(y_pred_3))
```

```python
m2 = model_multi.coef_[0]
b2 = model_multi.intercept_
print('y-=mx+b')
print(f'y = {m2: .0f}* + {b2: .0f}')
print(f'price  = {m2: .0f}*bedroom/bathroom + {b2: .0f}')
```

```python
def predict(bedrooms, bathrooms):
    y_pred3 = model_multi.predict([[bedrooms, bathrooms]])
    estimate3 = y_pred3[0]
    coefficient3 = model_multi.coef_[0]
    result3 = f'${estimate3:,.0f} esitmated price for {bedrooms:,.0f} bedroom {bathrooms:,.0f} bathrooms condo in New York. '
    explanation3 = f'In this linear regression, each additional bedroom/bathrooms adds ${coefficient3:,.0f}.'
    return result3 + explanation3
  
predict(4, 3)
```
