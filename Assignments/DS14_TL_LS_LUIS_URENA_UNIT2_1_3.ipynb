```python
import pandas as pd
import pandas_profiling

# Read New York City property sales data
df = pd.read_csv(DATA_PATH+'condos/NYC_Citywide_Rolling_Calendar_Sales.csv')

# Change column names: replace spaces with underscores
df.columns = [col.replace(' ', '_') for col in df]

# SALE_PRICE was read as strings.
# Remove symbols, convert to integer
df['SALE_PRICE'] = (
    df['SALE_PRICE']
    .str.replace('$','')
    .str.replace('-','')
    .str.replace(',','')
    .astype(int)
)
```

```python
# BOROUGH is a numeric column, but arguably should be a categorical feature,
# so convert it from a number to a string
df['BOROUGH'] = df['BOROUGH'].astype(str)
```

```python
# Reduce cardinality for NEIGHBORHOOD feature

# Get a list of the top 10 neighborhoods
top10 = df['NEIGHBORHOOD'].value_counts()[:10].index

# At locations where the neighborhood is NOT in the top 10, 
# replace the neighborhood with 'OTHER'
df.loc[~df['NEIGHBORHOOD'].isin(top10), 'NEIGHBORHOOD'] = 'OTHER'
```

```python
df = df[df['BUILDING_CLASS_CATEGORY'] == '01 ONE FAMILY DWELLINGS'] 
df.head()
```

```python
min_price = 100000
max_price = 2000000

df = df[df.SALE_PRICE >= min_price] 
```

```python
df.describe()
```

```python
df = df[df.SALE_PRICE < max_price]
```

```python
df['SALE_DATE'] = pd.to_datetime(df['SALE_DATE'],errors='ignore')
```

```python
df.dtypes
```

```python
train = df[df['SALE_DATE'].dt.month < 4]
test = df[df['SALE_DATE'].dt.month == 4]
```

```python
test
```

```python
target = 'SALE_PRICE'
high_cardinality = ['ADDRESS','LAND_SQUARE_FEET','BUILDING_CLASS_AT_PRESENT','BUILDING_CLASS_AT_TIME_OF_SALE','APARTMENT_NUMBER','EASE-MENT','SALE_DATE']
features = train.columns.drop([target] + high_cardinality)
```

```python
X_train = train[features]
Y_train = train[target]
X_test = test[features]
Y_test = test[target]
```

```python
import category_encoders as ce 
encoder  = ce.OneHotEncoder(use_cat_names=True)
X_train_encoded = encoder.fit_transform(X_train)
X_train_encoded
```

```python
from sklearn.linear_model import RidgeCV, Ridge
```

```python
rid_reg = Ridge()
```

```python
rid_reg.fit(X_train_encoded,Y_train)
```

```python
X_test_encoded = encoder.fit_transform(X_test)
```

```python
from sklearn.metrics import mean_absolute_error

y_pred = rid_reg.predict(X_test_encoded)
mae = mean_absolute_error(y_pred,Y_test)
```

```python
print(f'Test Mean Absolute Error: ${mae:,.0f} \n')
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.linear_model import RidgeCV, Ridge

```

```python
#instanciate the object pipeline
pipeline = make_pipeline(
    ce.OneHotEncoder(use_cat_names=True),
    SelectKBest(score_func=f_regression),
    Ridge(normalize=True)
    #RidgeCV()
)
```

```python
#fitting the model into the pipeline
pipeline.fit(X_train,Y_train)
```

```python
#pipeline test mean absolute error
y_pred = pipeline.predict(X_test)
mae = mean_absolute_error(y_pred,Y_test)

print(f'Test Mean Absolute Error(pipeline): ${mae:,.0f} \n')
```

```python
#new pipeline but with RidgeCV
pipeline = make_pipeline(
    ce.OneHotEncoder(use_cat_names=True),
    SelectKBest(score_func=f_regression),
    RidgeCV()
)
```

```python
pipeline.fit(X_train,Y_train)
```

```python
#RidgeCV Mean absolute error
y_pred = pipeline.predict(X_test)
mae = mean_absolute_error(y_pred,Y_test)

print(f'Test Mean Absolute Error(pipeline,RidgeCV): ${mae:,.0f} \n')
```
