```python
import pandas as pd
import pandas_profiling

# Read New York City property sales data
df = pd.read_csv(DATA_PATH+'condos/NYC_Citywide_Rolling_Calendar_Sales.csv')

# Change column names: replace spaces with underscores
df.columns = [col.replace(' ', '_') for col in df]

# SALE_PRICE was read as strings.
# Remove symbols, convert to integer
df['SALE_PRICE'] = (
    df['SALE_PRICE']
    .str.replace('$','')
    .str.replace('-','')
    .str.replace(',','')
    .astype(int)
)
```

```python
# BOROUGH is a numeric column, but arguably should be a categorical feature,
# so convert it from a number to a string
df['BOROUGH'] = df['BOROUGH'].astype(str)
```

```python
# Reduce cardinality for NEIGHBORHOOD feature

# Get a list of the top 10 neighborhoods
top10 = df['NEIGHBORHOOD'].value_counts()[:10].index

# At locations where the neighborhood is NOT in the top 10, 
# replace the neighborhood with 'OTHER'
df.loc[~df['NEIGHBORHOOD'].isin(top10), 'NEIGHBORHOOD'] = 'OTHER'
```

###  Use a subset of the data where BUILDING_CLASS_CATEGORY == '01 ONE FAMILY DWELLINGS' and the sale price was more than 100 thousand and less than 2 million.

```python
# Building a subset of the data

# refining down to Singly Family Dwellings
df_subset = df[df['BUILDING_CLASS_CATEGORY'] == '01 ONE FAMILY DWELLINGS']
```

```python
# refining further to a selling floor of $100,000
# and a selling ceiling of $2,000,000
df_subset = df_subset[df_subset['SALE_PRICE'] > 100000]
df_subset = df_subset[df_subset['SALE_PRICE'] < 2000000]
```

```python
df_subset.shape
```

### Do train/test split. Use data from January â€” March 2019 to train. Use data from April 2019 to test.

```python
df_subset['SALE_DATE'] = pd.to_datetime(df_subset['SALE_DATE'], infer_datetime_format=True)
split = pd.to_datetime('2019-04-01')

# splitting subset into train and test
train = df_subset[df_subset['SALE_DATE'] < split]
test = df_subset[df_subset['SALE_DATE'] >= split]
```

```python
df_subset.describe(exclude='number')
```

```python
# refine further into X & Y train and test datasets

# dropping categories with too many unique points to one hot encode
target = 'SALE_PRICE'
high_cardinality = ['ADDRESS', 'LAND_SQUARE_FEET', 'SALE_DATE', 'EASE-MENT']
features = train.columns.drop([target] + high_cardinality)
```

```python
X_train = train[features]
X_test = test[features]
y_train = train[target]
y_test = test[target]
```

### Do one-hot encoding of categorical features

```python
import category_encoders as ce 
encoder = ce.OneHotEncoder(use_cat_names=True)
X_train = encoder.fit_transform(X_train)
```

```python
X_test = encoder.transform(X_test)
```

```python
X_train.shape, X_test.shape
```

### Do feature selection with SelectKBest.

```python
X_train
```

```python
from sklearn.feature_selection import SelectKBest, f_regression

# defining how many features to select
selector = SelectKBest(score_func=f_regression, k=20)

X_train_selected = selector.fit_transform(X_train, y_train)
X_test_selected = selector.transform(X_test)
X_train_selected.shape, X_test_selected.shape
```

```python
# copy and pasted this code to find which features were selected
selected_mask = selector.get_support()
all_names = X_train.columns
selected_names = all_names[selected_mask]
unselected_names = all_names[~selected_mask]

print('Features selected:')
for name in selected_names:
    print(name)

print('\nFeatures not selected:')
for name in unselected_names:
    print(name)
```

```python
# using a function to determine how mnay features should be selected
# instead of hard coding a specific number

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

for k in range(1, len(X_train.columns)+1):
  print(f'{k} features')

  selector = SelectKBest(score_func=f_regression, k=k)
  X_train_selected = selector.fit_transform(X_train, y_train)
  X_test_selected = selector.transform(X_test)

  model = LinearRegression()
  model.fit(X_train_selected, y_train)
  y_pred = model.predict(X_test_selected)
  mae = mean_absolute_error(y_test, y_pred)
  print(f'Test Mean Absolute Error: ${mae:,.0f} \n')
```

### Fit a ridge regression model with multiple features. Use the normalize=True parameter 

```python
from sklearn.linear_model import Ridge, RidgeCV

# range of alphas to try to narrow down best range
# alphas = [0.0001, 0.001, 0.01, 0.1, 1.0, 1, 100.0, 1000.0]

# further narrowing range of best alphas
alphas = [0.009, 0.005, 0.001]
alphas_2 = [0.001, 0.002, 0.003, 0.004, 0.005, 0.006, 0.007, 0.008, 0.009]

ridge = RidgeCV(alphas=alphas_2, normalize=True)
ridge.fit(X_train, y_train)
ridge.alpha_
```

```python
# testing a much narrower range of alphas to determine best alpha

# are these testing different things? 

for alpha in [0.001, 0.002, 0.003, 0.004, 0.005, 0.006, 0.007, 0.008, 0.009]:
  print(f'Ridge Regression, with alpha={alpha}')
  model = Ridge(alpha=alpha, normalize=True)
  model.fit(X_train, y_train)
  y_pred = model.predict(X_test)

  # Get Test MAE
  mae = mean_absolute_error(y_test, y_pred)
  # display(HTML(f'Test Mean Absolute Error: ${mae:,.0f}'))
  print(f'Test Mean Absolute Error: ${mae:,.0f}')
```
